```python
#To plot standard normal distribution curve
def std_normaldist(data):
#finding mean and standard deviation for dataset
    mean=data.mean()
    std_D=data.std()

    
#creating a list for the range of column values
    values=[value for value in data]

#finding Z score with the formula
    zscore=[((j-mean)/std_D) for j in values]

#plot the normal distribution curve with distplot function
    sns.distplot(zscore,kde=True,kde_kws={'color':'blue'},color='grey')
```

```python
std_normaldist(data['degree_p'])
```
